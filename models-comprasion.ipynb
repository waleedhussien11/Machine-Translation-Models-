```python
from rouge import Rouge

def compute_rouge_scores(dataframe, reference_col, hypothesis_col):
    rouge = Rouge()

    rouge_scores = []

    # Iterate over rows and compute Rouge scores
    for index, row in dataframe.iterrows():
        reference = row[reference_col]
        hypothesis = row[hypothesis_col]

        # Compute Rouge scores
        scores = rouge.get_scores(hypothesis, reference)[0]

        # Extract the F1 score
        rouge_f1 = scores['rouge-1']['f']

        # Append scores to the list
        rouge_scores.append({'Row': index + 1, 'Rouge_F1': rouge_f1})

    # Create a DataFrame from the list of scores
    rouge_df = pd.DataFrame(rouge_scores)

    return rouge_df
```

# **BERT MODEL**

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")

```

```python

language = {
    'ar':'ar_AR',
    'ger':'de_DE','sp':'es_XX','fr':'fr_XX','ind':'hi_IN','ita':'it_IT','jap':'ja_XX','rus':'ru_RU','tur':'tr_TR','kor':'ko_KR','rom':'ro_RO'
}
```

```python
import torch
```

```python
def Translate(txt, lang, max_tokens=1024, length_penalty=2.0, num_beams=2):
    model_inputs = tokenizer(txt, return_tensors="pt")

    translation = model.generate(
        **model_inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id[lang],
        max_length=max_tokens,
        length_penalty=length_penalty,
        num_beams=num_beams
    )

    result = tokenizer.batch_decode(translation, skip_special_tokens=True)
    
    return str(result)
```

```python
result=Translate(df['en'][0],language['fr'])
```

```python
for i, row in df.iterrows():
    try:
        result = Translate(row['en'],language['fr'])
        df.at[i, 'bert'] = result
    except Exception as e:
        print(f"Null value  {i}: {e}")
    print(i)
```

```python
bert_score = compute_rouge_scores(df, 'fr', 'bert')
bert_score.mean()
```

```python
model_many = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt",src_lang="en_XX")
```

```python
def many(model,txt,lang):
    encoded_hi = tokenizer(txt, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded_hi,
        forced_bos_token_id=tokenizer.lang_code_to_id[lang],
        num_beams=3 , 
        max_length=1024
    )
    result= tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return str(result)
```

```python
from googletrans import Translator
def gog_trans(txt,src ,target,chunk_size=10000):
    translator = Translator()
    chunks = [txt[i:i+chunk_size] for i in range(0, len(txt), chunk_size)]

    chunk_tran = [translator.translate(chunk,src=src, dest=target).text for chunk in chunks]
    result = ' '.join(chunk_tran)
    return result
```

```python

for i, row in df.iterrows():
    try:
        result = gog_trans(row['en'],'en','fr')
        df.at[i, 'google']=result
    except Exception as e:
        print(f"Null value  {i}: {e}")
    print(i)
```

```python
def marian(model,txt): 
    model_input = tokenizer(txt, return_tensors="pt")

    generated_ids = model.generate(**model_input, max_length=4000,
                                  num_beams=3
                            )
    result= tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
    return str(result)
```
